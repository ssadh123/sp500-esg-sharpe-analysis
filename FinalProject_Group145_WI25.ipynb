



# Abstract

The motivation behind this study was to assess whether companies with stronger ESG risk scores exhibit better risk-adjusted performance, particularly during periods of market volatility. With increasing investor focus on ESG considerations, we aimed to determine if ESG factors provide a tangible financial advantage in uncertain market conditions.
To investigate this, we analyzed the Sharpe Ratios of S&P 500 companies alongside their ESG risk scores and market volatility (measured by the VIX) from 2020 to 2023. We conducted exploratory data analysis, correlation testing, and regression modeling to examine the relationship between ESG scores and risk-adjusted returns across different sectors and varying levels of market volatility.
Our findings indicate that there is no universal ESG premium—the relationship between ESG scores and Sharpe Ratios varies by sector. Communication Services and Energy sectors showed a positive correlation, suggesting that higher ESG scores may be linked to better performance in these industries. However, sectors like Basic Materials and Financial Services demonstrated a negative or weak correlation, indicating that ESG factors alone do not consistently drive superior returns. Regression analysis further confirmed that ESG scores and market volatility explain only a small portion of Sharpe Ratio variance, emphasizing the need for additional financial controls.
In conclusion, our results suggest that investors should take a sector-specific approach when incorporating ESG factors into investment decisions. While ESG may enhance performance in some industries, it does not universally improve risk-adjusted returns, particularly in volatile market conditions. Further research using panel data and alternative methodologies is needed to refine these insights and better understand ESG’s role in financial performance.

# Background and Prior Work


# ESG Factors and Financial Performance

Recent research highlights the growing impact of environmental, social, and governance (ESG) factors on firm performance. Khan, Serafeim, and Yoon (2016) found that firms with high sustainability scores tend to exhibit superior risk-adjusted returns, particularly when ESG factors are material to the industry. Their study provides evidence that ESG-conscious companies may have a competitive advantage, but it does not specifically address performance in high-volatility market environments. Similarly, Albuquerque, Koskinen, and Zhang (2020) argue that firms with strong ESG commitments experience lower systematic risk and tend to perform better during downturns. Aydoğmuş et al (2022) concludes that ESG performance (especially the social and governance aspects) has a significant, positive relationship with profitability and firm value. 

While it did not find a significant relationship with the environmental aspect, they suggested that environmental actions may take longer to affect profitability because of slower progress and higher costs. These findings suggest that companies with strong environmental risk management might be more resilient during volatile market conditions, aligning with our research focus.

---

# ESG Performance and Market Volatility (VIX > 30)

Other studies explore the relationship between ESG factors and market turbulence. Engle, Giglio, Kelly, Lee, and Stroebel (2020) demonstrated that climate-conscious firms are less exposed to downside risk, suggesting that ESG factors could serve as a hedge against market volatility. Meanwhile, Brière, Peillex, and Ureche-Rangau (2019) analyzed whether socially responsible investing (SRI) funds outperform during turbulent market conditions. Their results indicate that high-ESG funds tend to be more resilient but may underperform in certain volatility regimes, raising the question of whether environmental risk scores specifically influence risk-adjusted returns during extreme volatility periods. 

While these studies provide insights into ESG and volatility, they do not directly measure the interaction between environmental risk scores and stock performance under extreme volatility (VIX > 30), which is the central focus of our research.

**SP 500 ESG Risk Ratings** is a dataset with each of the S&P 500 companies and their corresponding Environmental, Social, and Governance (ESG) scores. ESG Risk Ratings are a numerical rating from 0-100 that is evaluate based on environmental risks: carbon footprint and greenhouse gas emissions, water usage, waste management, climate change risk, etc, social risks: labor practices, diversity and inclusion, human rights, community engagement, etc, and governance risks: executive compensation, business ethics and transparency, regulatory compliance, etc. Scores in each category are added together as weighted averages to form the Total ESG Risk Score. A higher rating indicates a better management of these risks and therefore more sustainable and ethical practices, while a lower rating suggests potential shortcomings or poor handling of ESG-related issues. The dataset includes a lot of other information that we do not need, so there are only a few columns that we care about for this project, ‘Name’ (of the company) and ‘Total ESG Risk Score', and the disaggregated values 'Environment Risk Score', 'Governance Risk Score', and 'Social Risk Score'. 
See https://portal.s1.spglobal.com/survey/documents/spglobal_esg_scores_methodology.pdf for more details on score formulation

---

**The CBOE VIX dataset** is using the Chicago Board Option Exchange for the Volatility Index, which is a measure of implied volatility and reflects how much investors anticipate the market to fluctuate in the near future (next ~30 days). Generally, a VIX value of over 30 is linked to a larger volatility and therefore increased uncertainty, investors’ fear, and risk while values below 20 correspond to more stable periods in the market. VIX is impacted by a variety of factors, including geopolitical events, market dynamics, and economic indicators (like GDP, inflation, interest rates, consumer spending, etc) as well as investor sentiment (e.g. positive market news can reduce fear, lowering the VIX), trading volumes, and option prices. The dataset shows what the VIX was at the end of each month from May 2020 to May 2023; we will disregard the other columns besides ‘Date’ and ‘Close’. 
We chose to use the close so that we were comparing the VIX at the same time each month and because the ESG scores, based on the Corporate Sustainability Assessment (series of questions that companies must answer, which is used to create the overall ESG score), are assessed every month—initial scores are released the third Friday of each month, so the Close is closest in time.

---

**The Sharpe Ratio dataset** shows the measure of the risk-adjusted return of an investment and helps to understand how well the return of an asset compensates for the risk taken. It is calculated with a formula that takes into account the average return of the investment, the risk-free rate (based on 3-Month Treasury Bill), and the standard deviation of the investment’s return. A high Sharpe Ratio (>1) would mean that the investment is providing a better return for the level of risk taken, which is generally preferred. A low Sharpe Ratio  (<1, or negative) means that the returns do not justify the level of risk and that the investment may not be a good choice. This is a useful way to compare different investments with varying levels of risk and return and allow investors to have more information when making decisions.

## Section 1: Testing for Correlations between Different Variables of the ESG and the AVG_Sharpe Ratio 

The below preliminary correlation check revealed that Environment, Social, and Governance scores are moderately correlated with one another, suggesting they might collectively influence risk‐adjusted returns more effectively than focusing on a single dimension. Based on these findings, we decided to **shift our question** from focusing solely on Environmental factors to using a combined Total ESG measure; the rest of our EDA will therefore focus on how average ESG risk relates to VIX and Sharpe ratios, rather than isolating the environmental component.


```python

# 1. READ THE S&P 500 ESG DATA (KEEPING ALL ROWS)

esg_file = "SP 500 ESG Risk Ratings.csv"
risk_ratings = pd.read_csv(esg_file)

# 2. READ THE SHARPE RATIO DATA (KEEPING ALL ROWS)
sharpe_file = "Sharpe Ratios for S&P 500 Co.csv"
df_sharpe = pd.read_csv(sharpe_file)

# 3. MERGE THE TWO DATASETS
merged_df = pd.merge(
    risk_ratings, 
    df_sharpe, 
    on="Symbol",
    how="inner"    
)

print("Merged dataset preview:")
display(merged_df.head())

# 4. ENCODE SECTOR TO INCLUDE IT IN THE CORRELATION MATRIX

if "Sector" in merged_df.columns:
    merged_df["Sector"] = merged_df["Sector"].astype(str)
    sectors = merged_df["Sector"].unique()
    sec_dict = {sec: i for i, sec in enumerate(sectors)}
    reverse_sec_dict = {i: sec for sec, i in sec_dict.items()}
    merged_df["Sector_Encoded"] = merged_df["Sector"].map(sec_dict)
else:
    print("No 'Sector' column found to encode.")

# 5. CREATE AN AVERAGE SHARPE RATIO COLUMN

sharpe_cols = ["Sept 22", "April 22", "Feb 22", "Jan 21", "Oct 20", "June 20"]
existing_sharpe_cols = [c for c in sharpe_cols if c in merged_df.columns]

if existing_sharpe_cols:
    merged_df["Avg_Sharpe"] = merged_df[existing_sharpe_cols].mean(axis=1, skipna=True)
else:
    print("No matching Sharpe columns found in the merged dataset.")

# 6. BUILD A CORRELATION MATRIX

num_cols = []
for col in [
    "Environment Risk Score",
    "Social Risk Score",
    "Governance Risk Score",
    "Total ESG Risk Score",
    "Avg_Sharpe",
    "Sector_Encoded"
]:
    if col in merged_df.columns:
        num_cols.append(col)

# Drop rows with missing data in these columns
df_for_corr = merged_df[num_cols].dropna()
corr_matrix = df_for_corr.corr()

print("\nCorrelation Matrix:")
display(corr_matrix)

# 7. VISUALIZE THE CORRELATION MATRIX WITH ANNOTATIONS INSIDE THE CELLS
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix for ESG Scores, Sector, and Sharpe Ratios")
plt.xticks(rotation=45, ha="right")
plt.show()

```

 This preliminary view helps identify which ESG factors might be more closely associated with risk‐adjusted returns, guiding further hypothesis testing.

## Section 2: Correlation Matrix for ESG Scores and Sharpe Ratios By Sectors 

**Brief Explanation of the Code**  
1. **Data Grouping**  
   - The dataset is grouped by sector so we can isolate and analyze each industry separately.

2. **Correlation Computation**  
   - For each sector, we compute a correlation matrix among the ESG risk scores (Environment, Social, Governance) and the Sharpe ratios (including average Sharpe ratio). This lets us see if, for instance, high Governance scores correlate with higher Sharpe ratios in the Industrials sector.

3. **Visualization**  
   - We use heatmaps (e.g., with Seaborn) to visualize each correlation matrix, making it easy to identify the strongest positive or negative relationships between ESG metrics and Sharpe ratios in each sector.

Overall, this approach clarifies whether ESG performance is equally relevant across different industries or if some sectors exhibit unique patterns worth investigating further.

### Sector Names

```python
print(merged_df[["Sector"]].head())
```

## Correlation Matrix by Sectors 

```python
sec = 'Sector_Encoded'
target_col = 'Avg_Sharpe'
cats = df_for_corr[sec].unique()
corr_matrices = {}

for cat in cats:
    cat_df = df_for_corr[df_for_corr[sec] == cat]
    cat_df = cat_df.drop(columns = [sec])
    
    corr_matrix = cat_df.corr()
    if target_col in corr_matrix.columns:
        target_corrs = corr_matrix[target_col].drop(target_col, errors='ignore')
        
        if np.any(np.abs(target_corrs) > 0.3):
            corr_matrices[cat] = corr_matrix

for cat, mat in corr_matrices.items():
    plt.figure(figsize=(8, 6))
    sns.heatmap(mat, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)

    plt.title(f"Correlation Matrix for {reverse_sec_dict[cat]} Sector companies ESG Scores and Sharpe Ratios\n (Correlation with Sharpe > 0.3)")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```

```python
model_sector = smf.ols("Q('Sharpe Ratio') ~ C(Sector) + Q('Close') * Q('Total ESG Risk score')", data=df).fit(cov_type='HC3')
print(model_sector.summary())
```

```python
# Variation Inflation Factor (VIF) to test for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

df = pd.get_dummies(df, columns=['Sector'], drop_first=True)

# Select your independent variables (predictors)
X = df[['Close', 'Total ESG Risk score', 'Sector_Communication Services', 'Sector_Consumer Cyclical',
       'Sector_Consumer Defensive', 'Sector_Energy',
       'Sector_Financial Services', 'Sector_Healthcare', 'Sector_Industrials',
       'Sector_Real Estate', 'Sector_Technology', 'Sector_Utilities']]

X = X.apply(lambda x: x.astype(float) if x.dtype == 'bool' else x)
X = X[np.isfinite(X).all(axis=1)]


# Add a constant to the model (intercept term)
X = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```

The OLS Regression has the following assumptions: linearity, autocorrelation, homoscedasticity, normality of errors, and no multicollinearity. 
- **Linearity:** we tested for linearity by graphing the Residuals against the Fitted values to see if there were any obvious patterns.
- **Autocorrelation:** the Durbin-Watson statistic that is returned with the OLS results was 2.115, which indicates no autocorrelation (independence of errors) so we meet that assumption.
- **Homoscedasticity:** by running a Breusch Pagan test, we test for homoscedasticity. Since the p-value from the test is very small, indicating there is heteroscedasticity (variance of errors is not constant), we address it by using Robust Standard Errors in our OLS Regression.
- **Normality of Errors:** the Jarque-Bera p-value from the OLS results was 0, meaning the residuals (errors) were non-normal. However, this should not be a problem since we have a large sample size that is robust to deviations and it is possible that our extreme outliers pushed the p-value towards significance (we did not want to remove the outliers and lose those datapoints).
- **Multicollinearity:** we ran a Variance Inflation Factor (VIF) to test the assumption of multicollinearity. The results show that the VIF is between 1 and 4 for all sectors, implying a moderate level of multicollinearity. Therefore, we should exhibit a level of caution since moderate multicollinearity can make it harder to interpret individual effects of predictors.
